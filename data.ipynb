```python
import pandas as pd

#############################################################
###   REVIEW THE 'stop' inconsistencies and fixing them   ###
#############################################################


### load the data into a df
economy_df = pd.read_csv('2/economy.csv')
business_df = pd.read_csv('2/business.csv')


###  Extracting the main values of the column 'stop'
economy_df['stop'] = economy_df['stop'].str.extract(r'(non-stop|1-stop|2-stop|2\+-stop)')
business_df['stop'] = business_df['stop'].str.extract(r'(non-stop|1-stop|2-stop|2\+-stop)')


#### Filling the null values
economy_df.fillna({'stop': 'unknown'}, inplace=True)
business_df.fillna({'stop': 'unknown'}, inplace=True)



### Printing the new values
print('INCONSISTENCIES DELETED FROM (stop)')
print(economy_df['stop'].value_counts())
print('business_df')
print(business_df['stop'].value_counts())



### SAVING THE FIRST ANOMLIE TO THE DATASET
economy_df.to_csv('2/economycleaned.csv', index=False)
business_df.to_csv('2/businesscleaned.csv', index=False)
```

```python
import re
###   SHOW THE ADVANCES TILL NOW
print('economy')
print(economy_df.head(1))
print('Business')
print(business_df.head(1))
print('######################################################')

print('\n GIVING HOUR FORMAT TO TIME TAKEN REMOVING H AND M')

def newformat(time_str):
    #Use regex to extract hour and minute
    match = re.match(r'(\d{2})h (\d{2})m', time_str)
    if match:
        hour = match.group(1)
        minute = match.group(2)
        return f"{hour}:{minute}"
    return time_str


economy_df['time_taken'] = economy_df['time_taken'].apply(newformat)
business_df['time_taken'] = business_df['time_taken'].apply(newformat)


print(economy_df['time_taken'].head(1))






```
